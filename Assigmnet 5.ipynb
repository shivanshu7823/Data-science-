(10)  Add a new column 'Score_Category': if Score >= 80 →
'Distinction', Score >= 60 → 'Merit', else 'Pass/Fail'. Use
np.where() or nested condition

```python
df["Score_Category"] = np.where(
    df["Score"]>=80,
    "Distinction",
    np.where(
        df["Score"]>=60,
        "Merit",
        "Pass/Fail"
    )
)

df
```
