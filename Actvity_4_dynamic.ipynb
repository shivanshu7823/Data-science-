```python
fig = px.sunburst(
    df,
    path=['Category', 'Product', 'City'],
    values='Sales_Amount',
    title='Sales Distribution by Category, Product and City'
)

fig.update_layout(
    title_x=0.5,
    width=900,
    height=700
)

fig.show()
```

```python
sankey_df = df.groupby(
    ['Category','Product']
)['Sales_Amount'].sum().reset_index()

sankey_df.head()
```

```python
labels = list(
    pd.concat([
        sankey_df['Category'],
        sankey_df['Product']
    ]).unique()
)

label_dict = {k:v for v,k in enumerate(labels)}

print(labels)
```

```python
source = sankey_df['Category'].map(label_dict)
target = sankey_df['Product'].map(label_dict)
value = sankey_df['Sales_Amount']
```

```python
fig = go.Figure(
    data=[
        go.Sankey(
            node=dict(
                pad=15,
                thickness=20,
                label=labels
            ),
            link=dict(
                source=source,
                target=target,
                value=value
            )
        )
    ]
)

fig.update_layout(
    title_text="Category → Product Sales Flow",
    title_x=0.5,
    width=1000,
    height=600
)

fig.show()
```

```python
new_row = pd.DataFrame({
    'Category':['Electronics'],
    'Product':['Laptop'],
    'City':['Delhi'],
    'Sales_Amount':[50000]
})

df = pd.concat([df,new_row], ignore_index=True)
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go


# SUNBURST
sunburst = px.sunburst(
    df,
    path=['Category','Product','City'],
    values='Sales_Amount',
    title='Sales Hierarchy Dashboard'
)

sunburst.update_layout(
    title_x=0.5,
    template='plotly_dark'
)

sunburst.show()

# SANKEY
sankey_df = df.groupby(
    ['Category','Product']
)['Sales_Amount'].sum().reset_index()

labels = list(
    pd.concat([
        sankey_df['Category'],
        sankey_df['Product']
    ]).unique()
)

label_dict = {k:v for v,k in enumerate(labels)}

source = sankey_df['Category'].map(label_dict)
target = sankey_df['Product'].map(label_dict)
value = sankey_df['Sales_Amount']

sankey = go.Figure(
    data=[
        go.Sankey(
            node=dict(
                pad=15,
                thickness=20,
                label=labels
            ),
            link=dict(
                source=source,
                target=target,
                value=value
            )
        )
    ]
)

sankey.update_layout(
    title='Category → Product Sales Flow',
    title_x=0.5,
    template='plotly_dark'
)

sankey.show()
```
